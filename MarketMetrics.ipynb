#### Cleaning missing values

```python
# Replacing missing review titles and messages with placeholders: "No title" and "No comment"

final_df['review_comment_title'] = final_df['review_comment_title'].fillna('No Title')
final_df['review_comment_message'] = final_df['review_comment_message'].fillna('No Comment')
final_df['review_comment_title'].head(10)
final_df['review_comment_message'].head(10)
```

```python
# Converting timestamp and date columns to datetime
timestamp_cols = [
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date',
    'shipping_limit_date',
    'review_answer_timestamp',
    'review_creation_date'
]

for col in timestamp_cols:
    final_df[col] = pd.to_datetime(final_df[col], errors='coerce')

final_df[col]
```

```python
# Dropping 726 rows where any of these key item-related columns are missing (having null values)

final_df_cleaned = final_df.dropna(
    subset=[
        'order_item_id', 'product_id', 'seller_id',
        'price', 'freight_value', 'shipping_limit_date'
    ]
)

# Checking shape change (expecting difference of 726 rows)
print(f"Original shape is", final_df.shape)
print(f"New shape after dropping: {final_df_cleaned.shape}")
```

```python
# Checking remaining null values

cleaned_missing_values_summary = final_df_cleaned.isnull().sum().to_frame(name='Missing Values')
cleaned_missing_values_summary['% Missing'] = (cleaned_missing_values_summary['Missing Values'] / len(final_df_cleaned)) * 100
cleaned_missing_values_summary = cleaned_missing_values_summary[cleaned_missing_values_summary['Missing Values'] > 0]

cleaned_missing_values_summary
```

```python
# Dropping rows where any of the single-row missing columns are null
single_row_missing_cols = [
    'payment_sequential', 'payment_type', 'payment_installments', 'payment_value',
    'review_id', 'review_score', 'review_creation_date', 'review_answer_timestamp'
]

final_df_cleaned = final_df_cleaned.dropna(subset=single_row_missing_cols)

# Confirming new shape
print("New shape after dropping single-row issues:", final_df_cleaned.shape)
```

```python
# Checking remaining null values

cleaned_missing_values_summary = final_df_cleaned.isnull().sum().to_frame(name='Missing Values')
cleaned_missing_values_summary['% Missing'] = (cleaned_missing_values_summary['Missing Values'] / len(final_df_cleaned)) * 100
cleaned_missing_values_summary = cleaned_missing_values_summary[cleaned_missing_values_summary['Missing Values'] > 0]

cleaned_missing_values_summary
```

```python
# Date/Time data cleaning

# For 'order_approved_at' (filling null values with median delay times)
delay = final_df_cleaned['order_approved_at'] - final_df_cleaned['order_purchase_timestamp']
print("Mean Delay:", delay.mean())
print("Median Delay:", delay.median())
# Using the median delay
median_delay = (
    final_df_cleaned['order_approved_at'] - final_df_cleaned['order_purchase_timestamp']
).median()

# Filling missing values using the median delay
final_df_cleaned['order_approved_at'] = final_df_cleaned['order_approved_at'].fillna(
    final_df_cleaned['order_purchase_timestamp'] + median_delay
)

final_df_cleaned['order_approved_at'].head(10)
```

```python
# Date/Time 
#For 'order_delivered_carrier_date'
# Calculating the time difference between carrier pickup and approval
carrier_delay = final_df_cleaned['order_delivered_carrier_date'] - final_df_cleaned['order_approved_at']

# Computing median delay
median_carrier_delay = carrier_delay.median()
print("Median Carrier Delay:", median_carrier_delay)

# Filling missing values using this delay
final_df_cleaned['order_delivered_carrier_date'].fillna(
    final_df_cleaned['order_approved_at'] + median_carrier_delay,
    inplace=True
)
```

```python
# Date/Time data cleaning

# Calculating delay between carrier pickup and customer delivery
delivery_delay = final_df_cleaned['order_delivered_customer_date'] - final_df_cleaned['order_delivered_carrier_date']

# Median delay
median_delivery_delay = delivery_delay.median()
print("Median Delivery Delay:", median_delivery_delay)

# Filling missing values
final_df_cleaned['order_delivered_customer_date'].fillna(
    final_df_cleaned['order_delivered_carrier_date'] + median_delivery_delay,
    inplace=True
)
```

```python
# Checking approval time fill
check_filled_approved = final_df_cleaned[
    final_df_cleaned['order_approved_at'] == final_df_cleaned['order_purchase_timestamp']
][['order_purchase_timestamp', 'order_approved_at']]
check_filled_approved
```

```python
# Checking carrier delay fill
carrier_delay = final_df_cleaned['order_delivered_carrier_date'] - final_df_cleaned['order_approved_at']
check_filled_carrier = final_df_cleaned[carrier_delay == pd.Timedelta("1 days 19:34:37")]
check_filled_carrier
```

```python
# Checking delivery delay fill
delivery_delay = final_df_cleaned['order_delivered_customer_date'] - final_df_cleaned['order_delivered_carrier_date']
check_filled_delivery = final_df_cleaned[delivery_delay == pd.Timedelta("7 days 02:16:30.5")]
check_filled_delivery
```

```python
# Checking remaining null values

cleaned_missing_values_summary = final_df_cleaned.isnull().sum().to_frame(name='Missing Values')
cleaned_missing_values_summary['% Missing'] = (cleaned_missing_values_summary['Missing Values'] / len(final_df_cleaned)) * 100
cleaned_missing_values_summary = cleaned_missing_values_summary[cleaned_missing_values_summary['Missing Values'] > 0]

cleaned_missing_values_summary
```

```python
# Dropping rows where ALL four metadata fields are missing
metadata_cols = [
    'product_category_name',
    'product_name_lenght',
    'product_description_lenght',
    'product_photos_qty'
]
mask_all_missing = final_df_cleaned[metadata_cols].isnull().all(axis=1)
final_df_cleaned = final_df_cleaned[~mask_all_missing]

# Imputing remaining missing values for fields not dropped
final_df_cleaned['product_category_name'].fillna('Unknown', inplace=True)
final_df_cleaned['product_name_lenght'].fillna(final_df_cleaned['product_name_lenght'].median(), inplace=True)
final_df_cleaned['product_description_lenght'].fillna(final_df_cleaned['product_description_lenght'].median(), inplace=True)
final_df_cleaned['product_photos_qty'].fillna(final_df_cleaned['product_photos_qty'].median(), inplace=True)

# Checking
original_rows = final_df_cleaned.shape[0]
dropped_rows = mask_all_missing.sum()  
print(f"Original rows: {original_rows}")
print(f"Rows with all 4 metadata columns missing: {dropped_rows}")
print(f"Remaining rows after drop: {original_rows - dropped_rows}")

print(final_df_cleaned[metadata_cols].isnull().sum())
```

### Exploratory Data Analysis

#### Delivery Delay Analysis: Checking how early or late an order was delivered relative to the estimated date

```python
# Creating delivery_delay column and assigning values to it
final_df_cleaned['delivery_delay'] = (
    final_df_cleaned['order_delivered_customer_date'] - final_df_cleaned['order_estimated_delivery_date']
).dt.days

final_df_cleaned['delivery_delay']
```

```python
final_df_cleaned[['order_delivered_customer_date', 'order_estimated_delivery_date', 'delivery_delay']].head(10)
```

```python
final_df_cleaned['delivery_delay'].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram

plt.figure(figsize=(10, 5))
sns.histplot(final_df_cleaned['delivery_delay'], bins=40, kde=True)
plt.title('Distribution of Delivery Delays (days)')
plt.xlabel('Delivery Delay (days)')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```

```python
# Checking percentage delay status by categories

def label_delay(days):
    if days > 0:
        return 'Late'
    elif days < 0:
        return 'Early'
    else:
        return 'On Time'

final_df_cleaned['delay_status'] = final_df_cleaned['delivery_delay'].apply(label_delay)

final_df_cleaned['delay_status'].value_counts(normalize=True) * 100
```

```python
#Pie Chart

# Counting the number of each delivery status
status_counts = final_df_cleaned['delay_status'].value_counts()

# Creating a pie chart with simple colors
plt.figure(figsize=(6, 6))
plt.pie(
    status_counts, 
    labels=status_counts.index, 
    autopct='%1.1f%%',  # Show percentages on the pie
    startangle=90,      # Rotate to make the first slice start at the top
    colors=['green', 'yellow', 'red']  # Easy-to-remember colors
)
plt.title('Delivery Status Distribution')
plt.axis('equal')  # Keep the pie chart as a circle
plt.show()
```

```python
# Bar Chart

# Counting the number of each delivery status
status_counts = final_df_cleaned['delay_status'].value_counts()

# Creating a bar chart
plt.figure(figsize=(8, 5))
plt.bar(status_counts.index, status_counts.values, color=['green', 'red', 'orange'])

# Adding chart title and labels
plt.title('Delivery Status Counts')
plt.xlabel('Delivery Status')
plt.ylabel('Number of Orders')

# Display the chart
plt.show()
```

#### Delivery Delay Analysis Summary

##### Objective:
A new column, `delivery_delay`, was created to calculate the difference in days between the actual delivery date and the estimated delivery date for each order.

##### Key Findings:
- The average delivery delay is −12 days, meaning most orders arrive earlier than expected.
- 50% of orders are delivered at least 13 days early.
- Only 6.6% of orders arrive later than estimated, and just 1.3% arrive exactly on time.

##### Anomalies and Observations:
- Some deliveries were up to 147 days early or 188 days late. These may indicate data entry errors or exceptional cases worth reviewing.
- The very low rate of on-time deliveries suggests that estimated delivery dates may be overly padded or misaligned with actual delivery operations.

##### Visualizations:
- A histogram was used to show the full distribution of delivery delays across all orders.
- A pie chart displays the proportion of Early, Late, and On-Time deliveries.
- A bar chart shows the number of orders in each delivery category.

##### Interpretation:
The company consistently delivers earlier than promised, which is a strong operational advantage. However, reviewing and adjusting estimated delivery windows may improve customer transparency and align expectations more accurately. Outliers should also be reviewed to ensure data quality and prevent reporting issues.


#### Delivery Time Analysis: Checking how long it took to deliver orders from purchase to custoemr delivery and cross-referencing trends by different factors.

```python
# Calculating delivery time (in days) as the difference between delivery date and purchase date and creating a new column for the values delivery_time
final_df_cleaned['delivery_time'] = (final_df_cleaned['order_delivered_customer_date'] - final_df_cleaned['order_purchase_timestamp']).dt.days
final_df_cleaned[['order_purchase_timestamp', 'order_delivered_customer_date', 'delivery_time']].head(10)
```

```python
final_df_cleaned['delivery_time'].describe()
```

```python
# Categorizing delivery speed based on quartiles from earlier summary
def categorize_speed(days):
    if days <= 5:
        return 'Fast'
    elif days <= 15:
        return 'Moderate'
    else:
        return 'Slow'

# Applying the function to create a new column
final_df_cleaned['delivery_speed'] = final_df_cleaned['delivery_time'].apply(categorize_speed)

final_df_cleaned['delivery_speed'].value_counts(normalize=True) * 100
```

```python
# Histogram for delivery_time

plt.figure(figsize=(10, 5))
plt.hist(final_df_cleaned['delivery_time'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Delivery Time (Days)')
plt.xlabel('Delivery Time (Days)')
plt.ylabel('Number of Orders')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Boxplot for delivery_time

plt.figure(figsize=(8, 4))
sns.boxplot(x=final_df_cleaned['delivery_time'], color='lightgreen')
plt.title('Boxplot of Delivery Time')
plt.xlabel('Delivery Time (Days)')
plt.tight_layout()
plt.show()
```

```python
# Bar chart for delivery_speed

speed_counts = final_df_cleaned['delivery_speed'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(speed_counts.index, speed_counts.values, color=['green', 'orange', 'red'])
plt.title('Order Counts by Delivery Speed')
plt.xlabel('Delivery Speed Category')
plt.ylabel('Number of Orders')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Pie chart for delivery_speed

speed_counts = final_df_cleaned['delivery_speed'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    speed_counts, 
    labels=speed_counts.index, 
    autopct='%1.1f%%', 
    startangle=90,
    colors=['orange', 'red', 'green']
)
plt.title('Delivery Speed Distribution')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

##### Trying to decipher the factors behind orders with fast delivery speeds and slow ones by considering location, product category, seller and season.

```python
# Counting of delivery speeds by state
state_speed = final_df_cleaned.groupby(['customer_state', 'delivery_speed']).size().unstack().fillna(0)

# Converting to percentages
state_speed_percent = state_speed.div(state_speed.sum(axis=1), axis=0) * 100

# Showing top 20 states by total orders
top_states = final_df_cleaned['customer_state'].value_counts().head(20).index
state_speed_percent.loc[top_states]
```

```python
# Stacked bar chart showing delivery_speed distribution by state

plt.figure(figsize=(14, 7))
state_speed_percent.plot(
    kind='bar',
    stacked=True,
    color=['green', 'orange', 'red'],
    figsize=(16, 7)
)

plt.title('Delivery Speed Distribution by State')
plt.xlabel('Customer State')
plt.ylabel('Percentage of Orders')
plt.legend(title='Delivery Speed')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
